```python
import pandas as pd
df = pd.read_csv('AQI_Data.csv')

#Display the first 5 rows 
print("Display first 8 rows : ")
print(df.head(8))

#Display last 8 rows 
print(df.tail(5))


```

```python
# Showing data type and number of non-null values in each column

print("Show data and number of non null values in each column :")
print(df.info())
print('\n')
```

```python
#Computing mean AQI , max PM2.5, min PM10 for each city

import numpy as np
print("Mean AQI, Max PM2.5, Min PM10 values for each day :")
print(df.groupby('Date').agg({'AQI': np.mean, 'PM2.5': np.max, 'PM10': np.min}))
```

```python
# QUestion 1.a

#Initialize an empty dictionary to store counts
row_counts = {}
#Loop through unique values in the 'City' column
for i in df['City'].unique():
    row_counts[i] = len(df[df['City'] == i])
#Print the count
print(row_counts)
```

```python
#Question 1.b

sum_pm2 = np.sum(df['PM2.5'])
print(sum_pm2)
sum_pm10 = np.sum(df['PM10'])
print(sum_pm10)
sum_no2 = np.sum(df['NO2'])
print(sum_no2)
sum_co = np.sum(df['CO'])
print(sum_co)
sum_o3 = np.sum(df['O3'])
print(sum_o3)
sum_so2 = np.sum(df['SO2'])
print(sum_so2)
```

```python
# Adding new column "Pollutant" to dataset
df['Pollutant'] = df[['PM2.5', 'PM10', 'NO2', 'CO', 'O3', 'SO2' ]].sum(axis=1)
df.drop("Sum", axis=1, inplace=False)
```

```python
# Saving it in a new file pollutant.txt
df.to_csv("pollutant.txt", sep="\t", index=False)

```
